```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_array(x: np.ndarray, filename: str) -> None:
    plt.clf()
    N = x.shape[0]
    idx = np.arange(0,N,1)

    plt.plot(idx, x)
    # plt.savefig(filename)
    plt.show()
    
```

```python
N_array = np.loadtxt("TotalRods.csv", delimiter=',')

```

```python
S_array = np.loadtxt("diffRods.csv", delimiter=',')
```

```python
plot_array(N_array, "")
```

```python
plot_array(S_array, "")
```
